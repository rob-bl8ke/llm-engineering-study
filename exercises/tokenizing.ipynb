```python
import tiktoken

encoding = tiktoken.encoding_for_model("gpt-4.1-mini")

tokens = encoding.encode("Hi my name is Rob and I like biltong")
```

Now that you have the tokens, you can display them.

```python
tokens
```

Turn the tokens back into words

```python
for token_id in tokens:
    token_text = encoding.decode([token_id])
    print(f"{token_id} = {token_text}")
```
